# Casos e bitos da COVID-19 no Brasil

# Objetivo do Estudo
Realizar uma anlise dos dados pblicos da pandemia da COVID-19 no Brasil, com enfoque nos registros de casos confirmados e bitos registrados devido  doena.

### 1.4 Valores Estatsticos

#### Mdia e Desvio Padro
- O Brasil teve uma mdia de 4213 casos confirmados por dia, com uma variao diria de 20 mil. Isso sugere que houve momentos de picos significativos de casos confirmados.
- Em mdia o Brasil registrou 110 bitos por dia devido  COVID-19. Assim como em casos confirmados, tambm houve uma variao nos bitos dirios, indicado pelo desvio padro de mais de 800 bitos. Isso sugere que houve momentos de picos significativos no nmero de bitos. 

#### Mdia e Mediana
-  possivel notar uma diferena entre a mediana e a mdia no resumo estatstico para as duas variveis. Essa diferena sugere a presena de outliers. Esses dados podem significar picos da COVID-19 em determinados perodos. Essa variao acentudada indica que o Brasil enfrentou altos nmeros de casos confirmados e de bitos registrada no perodo estudado do caso. 

### 1.5 Correlao

```python
# Criando uma matriz de correlao
df_corr = df.corr()

# Nmero de variveis a serem selecionadas
k = 6

# Encontrando as variveis mais correlacionadas com total_deaths
cols = df_corr.nlargest(k, 'deaths')['deaths'].index
cm = df[cols].corr()

# Plotando o heatmap
plt.figure(figsize=(10, 8))
hm = sns.heatmap(cm, cmap='Blues', cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 8}, yticklabels=cols.values, xticklabels=cols.values)
plt.title('Maiores Correlaes Positivas com bitos', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

A matriz de correlao mostra a relao linear entre diferentes variveis. Os valores variam de -1 a 1:
- 1: Correlao positiva perfeita (quando uma varivel aumenta, a outra tambm aumenta).
--1: Correlao negativa perfeita (quando uma varivel aumenta, a outra diminui).
- 0: Sem correlao (as variveis no tm relao linear).

#### Variveis
- confirmed e estimated_population (0.94): Regies mais populosas tendem a registrar mais casos confirmados e bitos.
- death_rate e confirmed_per_100k_inhabitants (-0.40): Existe uma correlao negativa moderada entre a taxa de mortalidade e a densidade de casos por 100 mil habitantes. Isso sugere que, em locais com maior densidade de casos, a taxa de mortalidade tende a ser menor.

#### Concluso
- O nmero de bitos est mais fortemente correlacionado com a populao estimada e o nmero de casos confirmados.
- Indicadores como a taxa de mortalidade (death_rate) e casos por 100 mil habitantes no esto diretamente relacionados ao nmero absoluto de bitos.
- A anlise sugere que regies mais populosas e com muitos casos confirmados concentram a maior parte dos bitos.

### 2. Anlise de Casos e bitos no Brasil


```python
#Agrupamento das Capitais com mais casos no total 
top_capitais_casos = df.groupby('state')['confirmed'].max().sort_values(ascending=False)[:10]
top_capitais_casos = top_capitais_casos.reset_index()
top_capitais_casos.columns = ['Estado', 'Total Casos']

display(top_capitais_casos)
```

```python
# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(12, 6))

# Criar o grfico de barras
bar_total_cases = ax.bar(top_capitais_casos['Estado'], top_capitais_casos['Total Casos'], color='lightblue', width=0.8)

# Formatar os nmeros
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# Funo para formatar valores em milhes
def formatar_milhares(x, pos):
    return f'{x / 1_0000:.0f}'

# Aplicar a formatao ao eixo Y
ax.yaxis.set_major_formatter(FuncFormatter(formatar_milhares))

# Remover grid 
ax.grid(False)

# Remover linhas
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Configurar rtulos do eixo y
ax.set_ylabel('Total de Casos (Em Milhes)', color='gray', fontsize=12, fontweight='bold')

# Definir cor cinza dos valores do eixo x
ax.tick_params(axis='x', colors='gray', labelsize=10)

# Definir cor cinza dos valores do eixo y
ax.tick_params(axis='y', colors='gray', labelsize=10)

# Remover yticks do eixo y
ax.tick_params(bottom=False)

# Adicionar os valores em cada barra, no mesmo formato
for bar in bar_total_cases:
    height = bar.get_height()
    ax.annotate(f'{height / 1_000:.0f} mil',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # Offset da posio do texto
                textcoords="offset points",
                ha='center', va='bottom', color='gray', fontsize=10)

# Destacar o maior valor
max_index = list(top_capitais_casos['Total Casos']).index(max(top_capitais_casos['Total Casos']))
for i, bar_total_cases in enumerate(bar_total_cases):
    if i != list(top_capitais_casos['Total Casos']).index(max(top_capitais_casos['Total Casos'])):
         bar_total_cases.set_color('lightgray')
            
# Ttulo
ax.text(-0.065, 1.20,'Dez Estados com Mais Casos Registrados de Covid-19',
        fontsize=20,
        color='gray',
        fontweight='bold',
        transform=ax.transAxes)

# Subttulo
ax.text(-0.065, 1.10,'Entre 2020 e 2023, os 10 estados brasileiras abaixo acumularam mais de'
                    ' 3.5 milhes de casos de COVID-19. \nDestaque para So Paulo (SP),'
                    ' que lidera a contagem de casos registrados do vrus.',
        fontsize=14,
        color='gray',
        transform=ax.transAxes);

# Legenda
ax.text(0.15, 0.80,'So Paulo (SP) totaliza mais de 950 mil de casos confirmados.\n'
                   'Esse valor corresponde  $\\bf26.27%$% do total entre os dez estados'
                    ' de casos confirmados.',
        fontsize=12,
        color='gray',
        transform=ax.transAxes);

# Ajustar espao entre os plots
plt.tight_layout()

# Apresentar o grfico
plt.show()
```

### 3. Anlise de Casos e por Habitantes

```python
# Agrupamento dos pases com mais casos por Habitantes
top_estados_casos_hab = df.groupby('state')['confirmed_per_100k_inhabitants'].max().sort_values(ascending=False)[:10]
top_estados_casos_hab = top_estados_casos_hab.reset_index()
top_estados_casos_hab.columns = ['Estado', 'Total de Casos por 100 Mil']
display(top_estados_casos_hab)

# Ordenar o DataFrame em ordem decrescente
top_estados_casos_hab = top_estados_casos_hab.sort_values(by='Total de Casos por 100 Mil', ascending=True)
```

```python
# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(12, 6))

# Criar o grfico de barras horizontais
bar_estados_casos_hab = ax.barh(top_estados_casos_hab['Estado'], top_estados_casos_hab['Total de Casos por 100 Mil'], color='lightblue')

# Formatar os nmeros sem notao cientfica
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# Remover grid
ax.grid(False)

# Remover linhas
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Definir cores das linhas
ax.spines['bottom'].set_color('gray')
ax.spines['left'].set_color('gray')

# Configurar rtulos do eixo x
ax.set_xlabel('Total de Casos por 100Mil de Habitantes', color='gray', fontsize=10, fontweight='bold')

# Definir cor cinza dos valores do eixo x
ax.tick_params(axis='x', colors='gray', labelsize=10)

# Definir cor cinza dos valores do eixo y
ax.tick_params(axis='y', colors='gray', labelsize=10)

# Remover ticks dos eixo y
ax.tick_params(left=False)

# Formatao dos rtulos de numeral completo para arredondado e milhares
def formatar_rotulo(valor):
    valor_arredondado = round(valor / 1000)
    return f'{valor_arredondado} Mil'

# Adicionar os valores em cada barra
for bar in bar_estados_casos_hab:
    width = bar.get_width()  # Obter o valor da barra para rtulo
    ax.annotate(f'{width / 1_000:.0f} mil',  # Formatando em milhares
                xy=(width, bar.get_y() + bar.get_height() / 2),  # Posio do rtulo ao final da barra
                xytext=(5, 0),  # Offset da posio do texto
                textcoords="offset points",
                ha='left', va='center', color='gray', fontsize=10)
    
# Ttulo
ax.text(-0.065, 1.20,'Dez estados com mais casos por Mil habitantes',
        fontsize=20,
        color='gray',
        fontweight='bold',
        transform=ax.transAxes);

# Subttulo
ax.text(-0.065, 1.10,'O grfico apresenta os 10 estados com as maiores taxas de casos de COVID-19 por Mil habitantes.\n'
                     'Destaque para Rio Grande do Norte (RN), que lidera a contagem de casos registrados do vrus por habitantes.',
        fontsize=14,
        color='gray',
        transform=ax.transAxes)

# Ajustar espao entre os plots
plt.tight_layout()

# Apresentar o grfico
plt.show()

```

### 4. Evoluo Temporal de Casos  

```python
# Converter a coluna 'date' para datetime, se ainda no estiver
df['date'] = pd.to_datetime(df['date'])

# Criar uma nova coluna 'year_month' com o ano e o ms apenas
df['year_month'] = df['date'].dt.to_period('M')

# Agrupar os dados por 'year_month' e somar os casos confirmados
df_agrupado_mes = df.groupby('year_month')['confirmed'].sum().reset_index()

# Ordenar o DataFrame por 'year_month' para garantir que os meses esto na ordem correta
df_agrupado_mes = df_agrupado_mes.sort_values(by='year_month')

# Exibir todos os resultados para confirmar a presena de dados de 2022
display(df_agrupado_mes)
```

```python
# Encontrar o ponto de maior valor
max_value = df_agrupado_mes['confirmed'].max()
max_date = df_agrupado_mes[df_agrupado_mes['confirmed'] == max_value]['year_month'].iloc[0]

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(16, 6), dpi=72)

# Certificar-se de que 'year_month' est em formato datetime
if not pd.api.types.is_datetime64_any_dtype(df_agrupado_mes['year_month']):
    df_agrupado_mes['year_month'] = pd.to_datetime(df_agrupado_mes['year_month'].astype(str))
       
# Criar o grfico de linha
ax.plot(df_agrupado_mes["year_month"], df_agrupado_mes["confirmed"], linestyle='-', color='darkblue');

# Formatar o eixo y para mostrar valores em milhares
def formatar_milhares(x, pos):
    return f'{x / 1_000000:.0f}'  # Dividir por mil para exibir em milhares

ax.yaxis.set_major_formatter(FuncFormatter(formatar_milhares))

# Definir os ticks para o eixo x por ms e formatar os rtulos
ax.xaxis.set_major_locator(mdates.MonthLocator())  # Marcadores a cada ms
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Formato: Jan 2020, Feb 2020, etc.

# Rotacionar e ajustar os rtulos para melhor legibilidade
plt.xticks(rotation=45)

# Remover as bordas superior e direita para um visual mais limpo
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Definir cores das linhas
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')

# Configurar rtulos do eixo y
ax.set_ylabel('Quantidade de Casos por Ms(Em Milhes)', color='gray', fontsize=12, fontweight='bold')

# Definir cor cinza dos valores do eixo x
ax.tick_params(axis='x', colors='gray', labelsize=10)

# Definir cor cinza dos valores do eixo y
ax.tick_params(axis='y', colors='gray', labelsize=10)

# Remover yticks do eixo y
ax.tick_params(left=False)

# Remover as linhas de grade 
ax.grid(False)

# Adicionar uma anotao para o maior valor
ax.annotate(
    'Maior Valor', 
    color='red', 
    xy=(pd.to_datetime(max_date), max_value),  # Ponto (data, valor)
    xytext=(pd.to_datetime(max_date), max_value + 1_000_000),  # Posio do texto
    arrowprops=dict(facecolor='red', shrink=0.05, width=1.5, headwidth=8)
)

# Ttulo
ax.text(-0.065, 1.20,'Evoluo de Casos de Covid-19 no Brasil entre 2021 e 2022',
        fontsize=20,
        color='gray',
        fontweight='bold',
        transform=ax.transAxes)

# Subttulo
ax.text(-0.065, 1.09,'Mais de 37 milhes de casos foram registrados no pas desde o incio da pandemia, '
                    'atingindo um pico de mais \nde 200 mil novos casos '
                    'por dia no incio de 2022.',
        fontsize=14,
        color='gray',
        transform=ax.transAxes)

# Legendas
ax.text(0.58, 0.90, 'Uma anlise feita com amostras positivas do SARS-Cov-2 \n'
                    'mostra que 31.7% das infeces foram causadas pela \n'
                    'variante micron entre os dias 1 e 15 de dezembro.',
        fontsize=12,
        color='gray',
        transform=ax.transAxes)

# Ajustar espao entre os plots
plt.tight_layout()

# Exibir o grfico
plt.show()
```
