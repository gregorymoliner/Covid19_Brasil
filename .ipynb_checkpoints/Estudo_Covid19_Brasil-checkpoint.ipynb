# Casos e Óbitos da COVID-19 no Brasil
O SARS-CoV-2 foi identificado pela primeira vez em seres humanos em dezembro de 2019 na cidade de Wuhan, na China. Pensa-se que o SARS-CoV-2 seja de origem animal.O surto inicial deu origem a uma pandemia global que à data de 15 de setembro de 2024.



# Objetivo do Estudo
Realizar uma análise dos dados públicos da pandemia da COVID-19 no Brasil, com enfoque nos registros de casos confirmados e óbitos registrados devido à doença.


# Obtenção dos Dados e Importação de Bibliotecas
Dados foram importados da Base de Dados Brasil.IO (https://brasil.io/covid19/)

### 1.4 Valores Estatísticos

```python
df[["confirmed", "deaths"]].describe().T
```

#### Média e Desvio Padrão
- O Brasil teve uma média de 4213 casos confirmados por dia, com uma variação diária de 20 mil. Isso sugere que houve momentos de picos significativos de casos confirmados.
- Em média o Brasil registrou 110 óbitos por dia devido à COVID-19. Assim como em casos confirmados, também houve uma variação nos óbitos diários, indicado pelo desvio padrão de mais de 800 óbitos. Isso sugere que houve momentos de picos significativos no número de óbitos. 

#### Média e Mediana
- É possivel notar uma diferença entre a mediana e a média no resumo estatístico para as duas variáveis. Essa diferença sugere a presença de outliers. Esses dados podem significar picos da COVID-19 em determinados períodos. Essa variação acentudada indica que o Brasil enfrentou altos números de casos confirmados e de óbitos registrada no período estudado do caso. 

### 1.5 Correlação

```python
# Criando uma matriz de correlação
df_corr = df.corr()

# Número de variáveis a serem selecionadas
k = 6

# Encontrando as variáveis mais correlacionadas com total_deaths
cols = df_corr.nlargest(k, 'deaths')['deaths'].index
cm = df[cols].corr()

# Plotando o heatmap
plt.figure(figsize=(10, 8))
hm = sns.heatmap(cm, cmap='Blues', cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 8}, yticklabels=cols.values, xticklabels=cols.values)
plt.title('Maiores Correlações Positivas com Óbitos', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

A matriz de correlação mostra a relação linear entre diferentes variáveis. Os valores variam de -1 a 1:
- 1: Correlação positiva perfeita (quando uma variável aumenta, a outra também aumenta).
--1: Correlação negativa perfeita (quando uma variável aumenta, a outra diminui).
- 0: Sem correlação (as variáveis não têm relação linear).

#### Variáveis
- confirmed e estimated_population (0.94): Regiões mais populosas tendem a registrar mais casos confirmados e óbitos.
- death_rate e confirmed_per_100k_inhabitants (-0.40): Existe uma correlação negativa moderada entre a taxa de mortalidade e a densidade de casos por 100 mil habitantes. Isso sugere que, em locais com maior densidade de casos, a taxa de mortalidade tende a ser menor.

#### Conclusão
- O número de óbitos está mais fortemente correlacionado com a população estimada e o número de casos confirmados.
- Indicadores como a taxa de mortalidade (death_rate) e casos por 100 mil habitantes não estão diretamente relacionados ao número absoluto de óbitos.
- A análise sugere que regiões mais populosas e com muitos casos confirmados concentram a maior parte dos óbitos.

### 2. Análise de Casos e Óbitos no Brasil


```python
#Agrupamento das Capitais com mais casos no total 
top_capitais_casos = df.groupby('state')['confirmed'].max().sort_values(ascending=False)[:10]
top_capitais_casos = top_capitais_casos.reset_index()
top_capitais_casos.columns = ['Estado', 'Total Casos']

display(top_capitais_casos)
```

```python
# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(12, 6))

# Criar o gráfico de barras
bar_total_cases = ax.bar(top_capitais_casos['Estado'], top_capitais_casos['Total Casos'], color='lightblue', width=0.8)

# Formatar os números
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# Função para formatar valores em milhões
def formatar_milhares(x, pos):
    return f'{x / 1_0000:.0f}'

# Aplicar a formatação ao eixo Y
ax.yaxis.set_major_formatter(FuncFormatter(formatar_milhares))

# Remover grid 
ax.grid(False)

# Remover linhas
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Configurar rótulos do eixo y
ax.set_ylabel('Total de Casos (Em Milhões)', color='gray', fontsize=12, fontweight='bold')

# Definir cor cinza dos valores do eixo x
ax.tick_params(axis='x', colors='gray', labelsize=10)

# Definir cor cinza dos valores do eixo y
ax.tick_params(axis='y', colors='gray', labelsize=10)

# Remover yticks do eixo y
ax.tick_params(bottom=False)

# Adicionar os valores em cada barra, no mesmo formato
for bar in bar_total_cases:
    height = bar.get_height()
    ax.annotate(f'{height / 1_000:.0f} mil',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # Offset da posição do texto
                textcoords="offset points",
                ha='center', va='bottom', color='gray', fontsize=10)

# Destacar o maior valor
max_index = list(top_capitais_casos['Total Casos']).index(max(top_capitais_casos['Total Casos']))
for i, bar_total_cases in enumerate(bar_total_cases):
    if i != list(top_capitais_casos['Total Casos']).index(max(top_capitais_casos['Total Casos'])):
         bar_total_cases.set_color('lightgray')
            
# Título
ax.text(-0.065, 1.20,'Dez Estados com Mais Casos Registrados de Covid-19',
        fontsize=20,
        color='gray',
        fontweight='bold',
        transform=ax.transAxes)

# Subtítulo
ax.text(-0.065, 1.10,'Entre 2020 e 2023, os 10 estados brasileiras abaixo acumularam mais de'
                    ' 3.5 milhões de casos de COVID-19. \nDestaque para São Paulo (SP),'
                    ' que lidera a contagem de casos registrados do vírus.',
        fontsize=14,
        color='gray',
        transform=ax.transAxes);

# Legenda
ax.text(0.15, 0.80,'São Paulo (SP) totaliza mais de 950 mil de casos confirmados.\n'
                   'Esse valor corresponde à $\\bf26.27%$% do total entre os dez estados'
                    ' de casos confirmados.',
        fontsize=12,
        color='gray',
        transform=ax.transAxes);

# Ajustar espaço entre os plots
plt.tight_layout()

# Apresentar o gráfico
plt.show()
```

### 3. Análise de Casos e por Habitantes

```python
# Agrupamento dos países com mais casos por Habitantes
top_estados_casos_hab = df.groupby('state')['confirmed_per_100k_inhabitants'].max().sort_values(ascending=False)[:10]
top_estados_casos_hab = top_estados_casos_hab.reset_index()
top_estados_casos_hab.columns = ['Estado', 'Total de Casos por 100 Mil']
display(top_estados_casos_hab)

# Ordenar o DataFrame em ordem decrescente
top_estados_casos_hab = top_estados_casos_hab.sort_values(by='Total de Casos por 100 Mil', ascending=True)
```

```python
# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(12, 6))

# Criar o gráfico de barras horizontais
bar_estados_casos_hab = ax.barh(top_estados_casos_hab['Estado'], top_estados_casos_hab['Total de Casos por 100 Mil'], color='lightblue')

# Formatar os números sem notação científica
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# Remover grid
ax.grid(False)

# Remover linhas
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Definir cores das linhas
ax.spines['bottom'].set_color('gray')
ax.spines['left'].set_color('gray')

# Configurar rótulos do eixo x
ax.set_xlabel('Total de Casos por 100Mil de Habitantes', color='gray', fontsize=10, fontweight='bold')

# Definir cor cinza dos valores do eixo x
ax.tick_params(axis='x', colors='gray', labelsize=10)

# Definir cor cinza dos valores do eixo y
ax.tick_params(axis='y', colors='gray', labelsize=10)

# Remover ticks dos eixo y
ax.tick_params(left=False)

# Formatação dos rótulos de numeral completo para arredondado e milhares
def formatar_rotulo(valor):
    valor_arredondado = round(valor / 1000)
    return f'{valor_arredondado} Mil'

# Adicionar os valores em cada barra
for bar in bar_estados_casos_hab:
    width = bar.get_width()  # Obter o valor da barra para rótulo
    ax.annotate(f'{width / 1_000:.0f} mil',  # Formatando em milhares
                xy=(width, bar.get_y() + bar.get_height() / 2),  # Posição do rótulo ao final da barra
                xytext=(5, 0),  # Offset da posição do texto
                textcoords="offset points",
                ha='left', va='center', color='gray', fontsize=10)
    
# Título
ax.text(-0.065, 1.20,'Dez estados com mais casos por Mil habitantes',
        fontsize=20,
        color='gray',
        fontweight='bold',
        transform=ax.transAxes);

# Subtítulo
ax.text(-0.065, 1.10,'O gráfico apresenta os 10 estados com as maiores taxas de casos de COVID-19 por Mil habitantes.\n'
                     'Destaque para Rio Grande do Norte (RN), que lidera a contagem de casos registrados do vírus por habitantes.',
        fontsize=14,
        color='gray',
        transform=ax.transAxes)

# Ajustar espaço entre os plots
plt.tight_layout()

# Apresentar o gráfico
plt.show()

```

### 4. Evolução Temporal de Casos  

```python
# Converter a coluna 'date' para datetime, se ainda não estiver
df['date'] = pd.to_datetime(df['date'])

# Criar uma nova coluna 'year_month' com o ano e o mês apenas
df['year_month'] = df['date'].dt.to_period('M')

# Agrupar os dados por 'year_month' e somar os casos confirmados
df_agrupado_mes = df.groupby('year_month')['confirmed'].sum().reset_index()

# Ordenar o DataFrame por 'year_month' para garantir que os meses estão na ordem correta
df_agrupado_mes = df_agrupado_mes.sort_values(by='year_month')

# Exibir todos os resultados para confirmar a presença de dados de 2022
display(df_agrupado_mes)
```

```python
# Encontrar o ponto de maior valor
max_value = df_agrupado_mes['confirmed'].max()
max_date = df_agrupado_mes[df_agrupado_mes['confirmed'] == max_value]['year_month'].iloc[0]

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(16, 6), dpi=72)

# Certificar-se de que 'year_month' está em formato datetime
if not pd.api.types.is_datetime64_any_dtype(df_agrupado_mes['year_month']):
    df_agrupado_mes['year_month'] = pd.to_datetime(df_agrupado_mes['year_month'].astype(str))
       
# Criar o gráfico de linha
ax.plot(df_agrupado_mes["year_month"], df_agrupado_mes["confirmed"], linestyle='-', color='darkblue');

# Formatar o eixo y para mostrar valores em milhares
def formatar_milhares(x, pos):
    return f'{x / 1_000000:.0f}'  # Dividir por mil para exibir em milhares

ax.yaxis.set_major_formatter(FuncFormatter(formatar_milhares))

# Definir os ticks para o eixo x por mês e formatar os rótulos
ax.xaxis.set_major_locator(mdates.MonthLocator())  # Marcadores a cada mês
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Formato: Jan 2020, Feb 2020, etc.

# Rotacionar e ajustar os rótulos para melhor legibilidade
plt.xticks(rotation=45)

# Remover as bordas superior e direita para um visual mais limpo
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Definir cores das linhas
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')

# Configurar rótulos do eixo y
ax.set_ylabel('Quantidade de Casos por Mês(Em Milhões)', color='gray', fontsize=12, fontweight='bold')

# Definir cor cinza dos valores do eixo x
ax.tick_params(axis='x', colors='gray', labelsize=10)

# Definir cor cinza dos valores do eixo y
ax.tick_params(axis='y', colors='gray', labelsize=10)

# Remover yticks do eixo y
ax.tick_params(left=False)

# Remover as linhas de grade 
ax.grid(False)

# Adicionar uma anotação para o maior valor
ax.annotate(
    'Maior Valor', 
    color='red', 
    xy=(pd.to_datetime(max_date), max_value),  # Ponto (data, valor)
    xytext=(pd.to_datetime(max_date), max_value + 1_000_000),  # Posição do texto
    arrowprops=dict(facecolor='red', shrink=0.05, width=1.5, headwidth=8)
)

# Título
ax.text(-0.065, 1.20,'Evolução de Casos de Covid-19 no Brasil entre 2021 e 2022',
        fontsize=20,
        color='gray',
        fontweight='bold',
        transform=ax.transAxes)

# Subtítulo
ax.text(-0.065, 1.09,'Mais de 37 milhões de casos foram registrados no país desde o início da pandemia, '
                    'atingindo um pico de mais \nde 200 mil novos casos '
                    'por dia no início de 2022.',
        fontsize=14,
        color='gray',
        transform=ax.transAxes)

# Legendas
ax.text(0.58, 0.90, 'Uma análise feita com amostras positivas do SARS-Cov-2 \n'
                    'mostra que 31.7% das infecções foram causadas pela \n'
                    'variante Ômicron entre os dias 1º e 15º de dezembro.',
        fontsize=12,
        color='gray',
        transform=ax.transAxes)

# Ajustar espaço entre os plots
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
